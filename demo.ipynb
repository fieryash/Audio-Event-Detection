# Demonstration Model

```python
# importing modules
import sounddevice as sd
import numpy as np
import librosa
import matplotlib.pyplot as plt
import librosa.display
from IPython.display import clear_output
from keras.models import load_model
import sklearn
```

```python
# loading the saved model
model=load_model("model.h5")
```

```python
# Setting up the sounddevice default values
sr=22050 #Sampling rate for recorder was set to 22.05 KHz
sd.default.samplerate=sr
sd.default.channels=1

# Recording an initial 30 seconds audio
duration=30.0000454
init_record=sd.rec(int(duration*sr))
sd.wait()
```

```python
# Function to record 5 seconds audio
def fun():
    mini_rec=sd.rec(sr*3)
    sd.wait()
    return mini_rec.reshape(-1)
```

```python
# Function to plot the MFCC each time
def plotter(fet,pred):
    if pred==0:
        title="babycry"
    elif pred==1:
        title="glassbreak"
    elif pred==2:
        title="gunshot"
    else:
        title="none"
    plt.figure(figsize=(15,5))
    librosa.display.specshow(fet,sr=sr,x_axis='time')
    plt.title(title)
    clear_output() # To clear the output screen
    plt.pause(0.001) # Wait to plot the MFCC
```

```python
# Loop to continuously record and predict the audio
import time
audio=init_record.reshape(-1)
audio_start = audio[sr: sr*15+1]
audio_end = audio[sr*17:-1]
while True:
    rec=fun()
    audio=np.concatenate((audio_start, rec, audio_end),axis=0)
    fet=sklearn.preprocessing.scale(librosa.feature.mfcc(audio,n_mfcc=40),axis=1)
    pred=np.argmax(model.predict(fet.reshape(1,fet.shape[0],fet.shape[1],1)))
    plotter(fet,pred)
    print(model.predict(fet.reshape(1,fet.shape[0],fet.shape[1],1)))
    print(time.ctime())
```
